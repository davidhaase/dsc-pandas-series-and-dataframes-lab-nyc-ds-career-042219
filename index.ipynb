## Compare subway traffic by day of the week. Display this as a graph.

```python
#Your code here
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('turnstile_180901.txt')
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
df['WEEKDAY'] = pd.to_datetime(df['DATE']).dt.dayofweek
day_counts = df.WEEKDAY.value_counts()
```

```python
ax = plt.subplot() # create an axis object, which the plot object which we can customize
x_day_names = [day_names[i] for i in range(0, len(day_counts))]
y_day_counts = [day_counts[i] for i in range(0, len(day_counts))]

plt.bar(range(len(x_day_names)), y_day_counts, color = 'purple', alpha = .5)

ax.set_xticks(range(0,len(x_day_names))) # set ticks values, as a method of the axes
ax.set_xticklabels(x_day_names) # set tick labels, also as a method of the axes
plt.ylim(27500, 29000)
plt.title('Count per Weekday')
plt.xlabel('Weekdays')
plt.ylabel('Counts')
plt.show()
```
